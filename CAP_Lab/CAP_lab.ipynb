# Empty Space


```python
file_dirs = '/content/drive/My Drive/CAP Lab/Measurements of Empty Space/.3 mm Measurements/*.csv'
files = glob.glob(file_dirs)

time = []
voltage = []
titlesN = []
titlesLN = []
titleN = 0
titleLN = 0


for curr_file in files:
  currentfile = pd.read_csv(curr_file)
  df = pd.DataFrame(currentfile, columns=['Time(s)', 'Ch 1  (V)'])
  dataset = df.values
  print(curr_file)

  lenarray = len(dataset)

  for i in range((lenarray - 76)):
    i = i+75
    x = float(dataset[i][0])
    time.append(x)
    y = float(dataset[i][1])
    voltage.append(y)
  
  plt.plot(time,voltage, '.')
  plt.show()
  plt.plot(time,np.log(voltage))
  plt.show()

  slopenintercept = np.polyfit(time,np.log(voltage),1)
  print(slopenintercept)

  time.clear()
  voltage.clear()



 
```
